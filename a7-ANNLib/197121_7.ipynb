#### 1. Implement a Simple Neural Network Model and predict if a person will buy insurance or not for the given dataset.

```python
# importing the required packages
import keras
from keras.models import Sequential   # For building the Neural Network layer by layer
from keras.layers import Dense        # used to add fully connected layer in ANN
import pandas as pd
import numpy as np
```

##### Loading the dataset


```python
# Load the dataset from the CSV file
import pandas as pd
dataset = pd.read_csv('insurance_data.csv')
dataset.head()
```

```python
# Normalising the data columns except species
dataset['Affordability'] = (dataset['Affordability'] - dataset['Affordability'].mean())/dataset['Affordability'].std()
dataset['age'] = (dataset['age'] - dataset['age'].mean())/dataset['age'].std()

dataset.head()
```

##### Splitting the dataset into X and Y

First split the data into Training and Testing dataset

```python
# Splitting the data into training and testing data
train_data = dataset.sample(frac=0.8, random_state=25)
test_data = dataset.drop(train_data.index)

# Printing the training and testing datasets shapes
print("The training dataset has {} rows and {} columns".format(train_data.shape[0], train_data.shape[1]))
print("The testing dataset has {} rows and {} columns".format(test_data.shape[0], test_data.shape[1]))

# Defining the training and testing data
X_train_insu = train_data[train_data.columns[:-1]]
Y_train_insu = train_data[train_data.columns[-1]]
X_test_insu = test_data[test_data.columns[:-1]]
Y_test_insu = test_data[test_data.columns[-1]]

# Printing the shapes of X and Y
print(X_train_insu.shape, X_test_insu.shape, Y_train_insu.shape, Y_test_insu.shape)
```

##### Generating the numpy arrays out of data

```python
# Generating the numpy arrays
X_train_insu = X_train_insu.values
Y_train_insu = Y_train_insu.values
X_test_insu = X_test_insu.values
Y_test_insu = Y_test_insu.values

# Reshaping train data into 2d matrices
Y_train_insu = Y_train_insu.reshape(len(Y_train_insu), 1)
Y_test_insu = Y_test_insu.reshape(len(Y_test_insu), 1)

print("Shapes of the matrices after splitting from the raw data")
print(X_train_insu.shape)
print(Y_train_insu.shape)
print(X_test_insu.shape)
print(Y_test_insu.shape)
```

##### Creating the network model using Keras

```python
# Initialise the model using Keras
model = Sequential()

# Dimensions
input_dim = X_train_insu.shape[1]
output_dim = 1

# Adding the input layer and the first hidden layer (of 16 neurons)
model.add(Dense(16, input_dim=2, activation='relu'))
# Adding the second hidden layer (of 12 neurons)
model.add(Dense(12, activation='relu'))
# Adding the output layer (of 1 neuron => Sigmoid)
model.add(Dense(1, activation='sigmoid'))
```

##### Now we need to specify the loss function and the optimizer. It is done using compile function in keras.

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

##### Training model

```python
model.fit(X_train_insu, Y_train_insu, epochs=100)
```

##### Predicting the test data labels

```python
# Finding the predictions from the model
pred = model.predict(X_test_insu)
for i in range(len(pred)):
    if pred[i][0] > 0.5:
        pred[i][0] = 1
    else:
        pred[i][0] = 0

print(pred)
```

##### Evaluating the accuracy of the model fit

```python
results = model.evaluate(X_test_insu, Y_test_insu)

print('Test loss:', results[0])
print('Test accuracy:', results[1])
```

#### 2. Churn Prediction model using Artificial Neural Networks

First import the dataset

```python
# Importing the dataset into the notebook
dataset = pd.read_csv('Churn_Modelling.csv')
dataset.head()
```

In the dataset, we donot need first three columns namely RowNumber, CustomerId, Surname. So let us drop them first

```python
# drop the unwanted columns
dataset = dataset.drop(dataset.columns[[0,1,2]], axis=1)
dataset.head()
```

Now we need to convert the string data (labels) in the dataset to numbers

```python
# Replacing country with numerical values
countries = dataset.iloc[:,1].unique()
contDict = {}
for i in range(len(countries)):
    contDict[countries[i]] = i

# Replacing gender with numerical values
genders = dataset.iloc[:, 2].unique()
gendDict = {}
for i in range(len(genders)):
    gendDict[genders[i]] = i



dataset = dataset.sample(frac=1, random_state=25)
dataset["Geography"] = dataset["Geography"].map(contDict)
dataset["Gender"] = dataset["Gender"].map(gendDict)
dataset.head()
```

Generate the numpy arrays for training and testing data for training the network model

```python
# Splitting the data into training and testing data
train_data = dataset.sample(frac=0.8, random_state=25)
test_data = dataset.drop(train_data.index)

# Printing the training and testing datasets shapes
print("The training dataset has {} rows and {} columns".format(train_data.shape[0], train_data.shape[1]))
print("The testing dataset has {} rows and {} columns".format(test_data.shape[0], test_data.shape[1]))

# Defining the training and testing data
X_train = train_data[train_data.columns[:-1]]
Y_train = train_data[train_data.columns[-1]]
X_test = test_data[test_data.columns[:-1]]
Y_test = test_data[test_data.columns[-1]]

# Printing the shapes of X and Y
print(X_train.shape, X_test.shape, Y_train.shape, Y_test.shape)
```

```python
# Generating the numpy arrays
# Generating the numpy arrays
X_train = X_train.values
Y_train = Y_train.values
X_test = X_test.values
Y_test = Y_test.values

# Reshaping train data into 2d matrices
Y_train = Y_train.reshape(len(Y_train), 1)
Y_test = Y_test.reshape(len(Y_test), 1)

print("Shapes of the matrices after splitting from the raw data")
print(X_train.shape)
print(Y_train.shape)
print(X_test.shape)
print(Y_test.shape)
```

##### Normalise the data 

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
X_train = sc.fit_transform(X_train) 
X_test = sc.fit_transform(X_test) 
```

#### Creating the ANN model using Keras

```python
# Initialise the model using Keras
model = Sequential()

# Dimensions
input_dim = X_train.shape[1]
output_dim = 1

# Adding the input layer and the first hidden layer (of 16 neurons)
model.add(Dense(16, input_dim=10, activation='relu'))
# Adding the second hidden layer (of 12 neurons)
model.add(Dense(12, activation='relu'))
# Adding the second hidden layer (of 12 neurons)
model.add(Dense(8, activation='relu'))
# Adding the output layer (of 1 neuron => Sigmoid)
model.add(Dense(1, activation='sigmoid'))
```

```python
# Compiling the model
model.compile(optimizer='adam', loss='binary_crossentropy',metrics=['accuracy'])
```

##### Train the model

```python
# Training the model
model.fit(X_train, Y_train, epochs=100)
```

```python
# Making predictions
# Finding the predictions from the model
pred = model.predict(X_test)
for i in range(len(pred)):
    if pred[i][0] >= 0.5:
        pred[i] = 1
    else:
        pred[i] = 0

print(pred)
```

```python
# Evaluationg model
results = model.evaluate(X_test, Y_test)

print('Test loss:', results[0]) 
print('Test accuracy:', results[1])
```
